```python
import pandas as pd
import pubchempy as pcp
```

```python
df_3d = pd.read_pickle('../output/bitter_sweet_2d_plus_3d_descriptors.pkl.gz', compression='gzip')
print(df_3d.shape)
```

```python
iupac_names = []
name = []
smiles = []
canonical_smiles = []
for i in range(df_3d.shape[0]):
    try:
        print("Try:", i)
        canc_ = df_3d.iloc[i]["Canonical SMILES"]
        s= pcp.get_compounds(canc_,'smiles')
        iupac_names.append(str(s[0].iupac_name))
        name.append(str(df_3d.iloc[i]["Name"]))
        smiles.append(str(df_3d.iloc[i]["SMILES"]))
        canonical_smiles.append(str(df_3d.iloc[i]["Canonical SMILES"]))
    except:
        print("Except:", i)
        iupac_names.append("Error")
        name.append(str(df_3d.iloc[i]["Name"]))
        smiles.append(str(df_3d.iloc[i]["SMILES"]))
        canonical_smiles.append(str(df_3d.iloc[i]["Canonical SMILES"]))
        continue
```

```python
df_iupac = pd.DataFrame({"Name": name, "IUPAC Names": iupac_names, "SMILES": smiles, "Canonical SMILES": canonical_smiles})
df_iupac
```

```python
df_iupac.to_csv('../output/3d-canon-smiles-iupac-names.tsv', sep='\t', index= False)
df_iupac.to_pickle('../output/3d-canon-smiles-iupac-names.pkl')
```
